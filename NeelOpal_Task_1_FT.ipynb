```python
import pandas as pd
```

```python
Data1 = pd.read_csv("/content/Records Matching Task.xlsx - data1.csv")
Data2 = pd.read_csv("/content/Records Matching Task.xlsx - data2.csv")
```

```python
Data1.head()
```

```python
Data2.head()
```

```python
# Identify records present in data1 but missing in data2
missing_records = Data1[~Data1[['Order ID', 'Product ID']].apply(tuple, axis=1).isin(Data2[['Order ID', 'Product ID']].apply(tuple, axis=1))]
```

```python
missing_records
```

```python
num_missing_records_1 = len(missing_records)
print(f"Number of records missing in data2: {num_missing_records_1}")
```

```python
# Identify records present in data1 but missing in data2
missing_records_2 = Data2[~Data2[['Order ID', 'Product ID']].apply(tuple, axis=1).isin(Data1[['Order ID', 'Product ID']].apply(tuple, axis=1))]

num_missing_records_2 = len(missing_records_2)
```

```python
print(f"Number of records missing in data1: {num_missing_records_2}")
```

```python
print(f"Totoal number of records missing : {num_missing_records_2 + num_missing_records_1}")
```

```python
combined_data = pd.concat([Data1, Data2])

Unique_records = combined_data[['Order ID', 'Product ID']].drop_duplicates()
```

```python
num_unique = Unique_records.shape[0]

print(f"Total number of unique records = {num_unique}")
```
